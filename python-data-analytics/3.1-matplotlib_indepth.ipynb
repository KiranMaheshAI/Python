```python
import matplotlib.pyplot as plt
import pandas as pd

x= [1,2,3]
y= [4,5,6]

plt.plot(x,y) # (1,4) (2,5) (3,6)
plt.grid()  # show grid way
plt.show()
```

```python
# Pyplot API
# Univariate -  Numerical
data = {
    "salary": [100, 200, 50, 400, 90]
}
df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
# Line Plot
plt.plot(df["salary"], color="red", marker="*", linestyle = ":", linewidth="2")
plt.grid()
plt.show()
```

```python
# Histogram

plt.hist(df["salary"], color="green",bins=5)
plt.show()
```

```python
# Box plot
plt.boxplot(df["salary"])
plt.show()
```

```python
df.loc[20] = 0
```

```python
# Box plot
plt.boxplot(df["salary"])
plt.show()
```

```python
df.drop(index=20, inplace=True)
```

```python
# Univariate: Categorical

df["dept"] = ["HR", "IT", "Finance", "HR", "IT"]
df
```

```python
df.info()
```

```python
# Pie Chart
count = df["dept"].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct="%1.1f", explode=[0, 0.1, 0.2])
plt.axis("equal")
plt.show()
```

```python
# Countplot
plt.bar(count.index, count, color = ["green", "red", "yellow"])
plt.show()
```

```python
# Bivariate  - numerical - numerical
df.head()
```

```python
df["age"] = [10, 15, 20, 25, 30]
df.head()
```

```python
# scatter plot
plt.scatter(df["age"], df["salary"], color="orange")
plt.show()
```

```python
# Line plot
plt.plot(df["age"], df["salary"], color="red", marker="*", linestyle=":", linewidth="2")
plt.grid()
plt.show()
```

```python
# Bar Chart

plt.bar(df["salary"], df["age"], color="green", align="edge")
plt.show()
```

```python
df
```

```python
# Bivariate : Numerical - Categorical

hr_sal = df[df["dept"] == "HR"]["salary"]
print(hr_sal)

it_sal = df[df["dept"] == "IT"]["salary"]
print(it_sal)

fin_sal = df[df["dept"] == "Finance"]["salary"]
print(fin_sal)
```

```python
# Boxplot
plt.boxplot([hr_sal, it_sal, fin_sal], labels=["HR", "IT", "Finance"])
plt.legend()
plt.show()
```

```python
# pie chart
salary_by_dept = df.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels=salary_by_dept.index, autopct="%1.2f", shadow=True, explode=[0.1, 0.2, 0.3])
plt.axis("equal")
plt.show()
```

```python
# Bar plot
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(hr_sal)
fin_mean = sum(fin_sal)/len(fin_sal)

plt.bar(["HR", "IT", "Finance"], [hr_mean, it_mean, fin_mean], color = ["green", "black", "red"])
plt.show()
```

```python
# Multivariate Analysis: 3 numerical columns
df
```

```python
df["experience"] = [3, 4, 1, 5, 2]
df
```

```python
# Bubble plot : Larger the bubble, more the experience they have.

plt.scatter(df["age"], df["salary"], s=df["experience"]*20, color="orange", edgecolors="black")
plt.title(" Age vs Salary vs Experience ")
plt.xlabel("Age")
plt.ylabel("salary")
plt.show()
```

```python
# 2 Numerical and 1 Categorical

plt.scatter(df["age"], df["salary"], c=df["dept"].map({"HR": "yellow", "IT": "blue", "Finance": "orange"}))
plt.xlabel("Age")
plt.ylabel("salary")
plt.show()
```

```python
# LEGENDS

color = {"HR": "yellow", "IT": "red", "Finance": "blue"}

for dept, color in color.items():
    df_dept = df[df["dept"]==dept]
    plt.scatter(df_dept["age"], df_dept["salary"], c=color, label=dept)

plt.legend()
```

### Object Oriented API

```python
fig, axs = plt.subplots(1, 3, figsize = (10, 5))

# Line Plot
axs[0].plot(df["age"], df["salary"], color="red", marker="*", linewidth="2", markersize=2)
axs[0].grid()
axs[0].set_title("Line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("salary")

# histogram
axs[1].hist(df["salary"], bins=5, color="skyblue")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")

# Box plot
axs[2].boxplot(df["salary"])
axs[2].set_title("Box plot")
axs[2].set_xlabel("Salary")
axs[2].set_ylabel("Frequency")

plt.savefig("multiplePlots.png")
plt.show()
```

```python
from pandas import DataFrame


data2 = {
    'year': [2020, 2021, 2022, 2023],
    'sales': [100, 150, 200, 250],
    'profit': [20, 30, 40, 50],
    'expenses': [80, 120, 160, 200]
}

df2 = pd.DataFrame(data2)

plt.plot(df2['year'], df2["sales"], label='Sales')
plt.plot(df2['year'], df2["profit"], label='Profit')
plt.plot(df2['year'], df2["expenses"], label='Expenses')

plt.title("Finance analysis")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

## 3D Plot

```python
ax = plt.axes(projection="3d")
ax.scatter(df["age"], df["salary"], df["experience"])
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")
plt.show()
```
