```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

```python
sns.set_style("whitegrid") # darkgrid, dark, ticks
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="island")
sns.despine(left=True) # remove the left vertical line
```

```python
sns.set_style("ticks")
sns.set_context('paper')
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="island")
sns.despine(left=True)
```

```python
sns.set_style("ticks")
sns.set_context('notebook') # Default context
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="island", palette="Set3") # Palette is Set3
sns.despine(left=True)
```

```python
sns.scatterplot(data=penguins, x="species", y="body_mass_g", hue="island", style="sex")
```

```python
# Strip Plot
sns.stripplot(data=penguins, x="species", y="body_mass_g", hue="island")
```

```python
# Strip Plot with dodge and jitter
sns.stripplot(data=penguins, x="species", y="body_mass_g", hue="island", dodge=True, jitter=True)
```

```python
# Swarm Plot
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island")
```

```python
sns.set_context('talk')
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island")
sns.despine()
```

```python
# Histogram
sns.set_context("notebook")
sns.histplot(data=penguins, x="body_mass_g")
```

```python
# Histogram
sns.set_context("notebook")
sns.histplot(data=penguins, x="body_mass_g", hue="sex", multiple="stack")
```

```python
# Regression Plot
sns.regplot(data=penguins, x="body_mass_g", y="flipper_length_mm", color="green")
```

```python
# Regression Plot
sns.regplot(data=penguins, x="body_mass_g", y="flipper_length_mm", scatter=False)
```

```python
# line Plot
sns.set_style("whitegrid")
sns.lineplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="island", style="sex")
```

```python
# joint plots
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="sex")
```

```python
# joint plots
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue="island", kind="scatter")
```

```python
# joint plots
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", kind="reg")
```

```python
# Bar plot
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["pink", "skyblue"])
```

```python
# Bar plot with estimator `sum`
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["pink", "skyblue"], estimator=np.sum)
```

```python
# Bar plot with estimator `varience`
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["pink", "skyblue"], estimator=np.var)
```

```python
# count plot
sns.countplot(data=penguins, x="species", hue="island", palette="Set2")
```

```python
# Box plot with island hue

sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="island")
```

```python
# Box plot with sex hue

sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="sex")
```

```python
# violin plot
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex")
```

```python
# violin plot
# Width of violin: Density of data at different values
# Middle line : Median
# Thick bar: Interquartile range 

sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex", split=True)
```

```python
sns.violinplot(data=penguins, x="species", y="body_mass_g")
sns.swarmplot(data=penguins, x="species", y="body_mass_g", color="black", size=3) 
```

```python
# KDE stands for estimating the probability density function of continuous or non-parametric data
# In simple terms, smooth curve that shows the distribution of data - like a smoothed-out version
# It helps you see where the data is concentrated (peaks) and how it spreads

sns.kdeplot(data=penguins, x="body_mass_g", hue="species", palette="pastel", fill=False) 
```

```python
# HeatMap: a 2D color coded matrix that shows relationships or patterns between two variables
# It's commonly used to:
#   - Show correlation between features
#   - Visualize confusion matrices - Will see in later stages
#   - Highlight missing data -> sns.heatmap()
```

```python
penguins.head()
```

```python
columns =["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
penguins[columns].corr()
```

```python

sns.heatmap(data=penguins[columns].corr(), annot=True, cmap="Blues", vmin=-0.2, linewidth=2, linecolor="black")
```

```python
sns.heatmap(data=penguins[columns].corr(), annot=True, cmap="Blues")
```

```python
# Rug Plot:
sns.rugplot(data=penguins, x="body_mass_g", hue="species", palette="Set2")
```

```python
sns.rugplot(data=penguins, x="body_mass_g", hue="species", palette="Set2", height=0.2)
```

```python
# Pair plot
sns.pairplot(data=penguins, hue="sex", palette="Set2")
```

```python
# Pair plot
sns.pairplot(data=penguins[columns])
```

```python
# Diagonal kind indicates histogram 
sns.pairplot(data=penguins, hue="sex", palette="Set2", diag_kind="hist")
```

```python
# Diagonal kind indicates kde 
sns.pairplot(data=penguins, hue="sex", palette="Set2", diag_kind="kde")
```

```python
# Pair grid -> subplot grid for plotting pairwise relationship in a dataset
graph=sns.PairGrid(data=penguins, hue="sex", palette="Set2")
graph.map_upper(sns.scatterplot)  # Upper triangle
graph.map_lower(sns.kdeplot)  # Upper triangle
graph.map_diag(sns.histplot)  # Diagonal
graph.add_legend()
```
